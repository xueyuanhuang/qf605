## Market implied vol with different maturity

### Function for Black-Scholes

```python
#black-sholes
def BlackScholesLognormalCall(S, K, r, sigma, T):
    d1 = (np.log(S/K)+(r+sigma**2/2)*T) / (sigma*np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)
    return S*norm.cdf(d1) - K*np.exp(-r*T)*norm.cdf(d2)


def BlackScholesLognormalPut(S, K, r, sigma, T):
    d1 = (np.log(S/K)+(r+sigma**2/2)*T) / (sigma*np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)
    return K*np.exp(-r*T)*norm.cdf(-d2) - S*norm.cdf(-d1)
```

### Function for calculate implied volatility

```python
#market implied vol
def impliedVolatility_bs(S, K, r, price, T, payoff):
    try:
        if (payoff.lower() == 'call'):
            impliedVol = brentq(lambda x: price -
                                BlackScholesLognormalCall(S, K, r, x, T),
                                1e-12, 1000.0)
        elif (payoff.lower() == 'put'):
            impliedVol = brentq(lambda x: price -
                                BlackScholesLognormalPut(S, K, r, x, T),
                                1e-12, 1000.0)
        else:
            raise NameError('Payoff type not recognized')
    except Exception:
        impliedVol = np.nan

    return impliedVol
```

### Use OTM-PUT and OTM-CALL to gain the volatility smile

```python
def transfer_to_implied_vol(df,S):

    call_df = df[df['payoff'] == 'call']
    put_df = df[df['payoff'] == 'put']
    strikes = put_df['strike'].values
    impliedvols = []
    payoff = []
    price = []
    for K in strikes:    
        if K > S:
            impliedvols.append(call_df[call_df['strike'] == K]['vols_bs'].values[0])
            payoff.append(call_df[call_df['strike'] == K]['payoff'].values[0])
            price.append(call_df[call_df['strike'] == K]['mid_price'].values[0])
        else:
            impliedvols.append(put_df[put_df['strike'] == K]['vols_bs'].values[0])
            payoff.append(put_df[put_df['strike'] == K]['payoff'].values[0])
            price.append(put_df[put_df['strike'] == K]['mid_price'].values[0])

# populate "df" with the dataframe containing strikes and market implied volatilities
    df = pd.DataFrame({'strike': strikes, 'impliedvol': impliedvols, "payoff": payoff, "price": price})
    return df,strikes
```

## Displaced-Diffusion model

>Step1: Calculate ATM sigma by using BS model

```python
def calc_sigma(df1,S,r,T):
    min_difference = np.abs( df1["strike"].iloc[0] - S )
    for i in (df1[["strike"]].index):
        diff = np.abs( df1['strike'].loc[i] - S )
        if diff <= min_difference:
            min_difference = diff
            closest_to_s0 = df1['mid_price'].loc[i]
            type_closest_to_s0 = df1["payoff"].loc[i]
        
            ATM_volatility = impliedVolatility_bs(S, S, r, closest_to_s0, T, type_closest_to_s0)
            
    return ATM_volatility
```

```python
sigma1_spx = calc_sigma(spx_df1,Sx,r1,T1)
sigma2_spx = calc_sigma(spx_df2,Sx,r2,T2)
sigma3_spx = calc_sigma(spx_df3,Sx,r3,T3)

sigma1_spy = calc_sigma(spy_df1,Sy,r1,T1)
sigma2_spy = calc_sigma(spy_df2,Sy,r2,T2)
sigma3_spy = calc_sigma(spy_df3,Sy,r3,T3)
```

>step2: Estimate best beta

```python
def black76_Call(F, K, T, r, sigma):
    discount_factor = np.exp(-r*T)
    #F = S0*np.exp(r*T)
    d1 = (np.log(F/K)+(1/2)*(sigma**2)*T)/(sigma*np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    price = discount_factor * ((F*norm.cdf(d1)) - (K*norm.cdf(d2)))
    return price
def black76_Put(F, K, T, r, sigma):
    discount_factor = np.exp(-r*T)
    #F = S0*np.exp(r*T)
    d1 = (np.log(F/K)+1/2*(sigma**2)*T)/(sigma*np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    price = discount_factor * ((K*norm.cdf(-d2)) - (F*norm.cdf(-d1)))
    return price
```

```python
def dd_calc_price(r, T, S, K, beta, sigma, payoff):
    F = S*np.exp(r*T)
    if (payoff.lower() == 'call'):
        return(black76_Call(F/beta, K+((1-beta)/beta)*F, T, r, sigma*beta))
    elif (payoff.lower() == 'put'):
        return(black76_Put(F/beta, K+((1-beta)/beta)*F, T, r, sigma*beta,))
```

```python
def ddcalibration(x, strikes, vols, S, r, T, sigma, payoff):
    err = 0.0
    for i, vol in enumerate(vols):
        price = dd_calc_price(r, T, S, strikes[i], x[0], sigma, payoff[i])
        err += (vol - impliedVolatility_bs(S, strikes[i], r, price, T, payoff[i]))**2
    return err
```

```python
def DD_fit_beta(strike, implied_vol, S, r, T, sigma, payoff):
    initialGuess = [0.7]
    bounds = (0,1)
    res = least_squares(lambda x: ddcalibration(x,
                                              strike,
                                              implied_vol,
                                              S, r,
                                              T,sigma, payoff),
                    initialGuess, bounds = bounds)
    return res.x[0]
```

```python
beta1_spx = DD_fit_beta(spx_df1_bs['strike'], spx_df1_bs['impliedvol'], Sx, r1, T1, sigma1_spx, spx_df1_bs['payoff'])
beta2_spx = DD_fit_beta(spx_df2_bs['strike'], spx_df2_bs['impliedvol'], Sx, r2, T2, sigma2_spx, spx_df2_bs['payoff'])
beta3_spx = DD_fit_beta(spx_df3_bs['strike'], spx_df3_bs['impliedvol'], Sx, r3, T3, sigma3_spx, spx_df3_bs['payoff'])

beta1_spy = DD_fit_beta(spy_df1_bs['strike'], spy_df1_bs['impliedvol'], Sy, r1, T1, sigma1_spy, spy_df1_bs['payoff'])
beta2_spy = DD_fit_beta(spy_df2_bs['strike'], spy_df2_bs['impliedvol'], Sy, r2, T2, sigma2_spy, spy_df2_bs['payoff'])
beta3_spy = DD_fit_beta(spy_df3_bs['strike'], spy_df3_bs['impliedvol'], Sy, r3, T3, sigma3_spy, spy_df3_bs['payoff'])
```

```python
def renew_data(string): #SPX or SPY
    df = pd.read_csv(string+'_options.csv')
    df["mid_price"] = (df["best_bid"]+df["best_offer"])/2
    df['strike'] = df['strike_price']*0.001
    df['payoff'] = df['cp_flag'].map(lambda x: 'call' if x == 'C' else 'put')

    exdate1 = sorted(df['exdate'].unique())[0]
    exdate2 = sorted(df['exdate'].unique())[1]
    exdate3 = sorted(df['exdate'].unique())[2]

    df1 = df[df['exdate'] == exdate1]
    df2 = df[df['exdate'] == exdate2]
    df3 = df[df['exdate'] == exdate3]
    return df1, df2, df3
```

```python
def calc_dd_vol(df,r,T,S,beta,sigma):
    df["price_dd"] = df.apply(lambda x: dd_calc_price (r, 
                                                             T, 
                                                             S, 
                                                             x["strike"], 
                                                             beta, 
                                                             sigma, 
                                                             x["payoff"]),axis=1)
    df['vols_dd'] = df.apply(lambda x: impliedVolatility_bs(S,
                                                  x["strike"],
                                                  r,
                                                  x['price_dd'],
                                                  T,
                                                  x["payoff"]),axis=1)
    return df
```

## SABR Model

>Step1: Define function to estimate parameters in SABR function

```python
def sabrcalibration(x, strikes, vols, F, T, beta):     #x is a df there
    err = 0.0
    for i, vol in enumerate(vols):
        err += (vol - SABR(F, strikes[i], T,
                           x[0], beta, x[1], x[2]))**2
    return err
```

>Step2: Define function to calculate implied volatility by using SABR

```python
def SABR(F, K, T, alpha, beta, rho, nu):
    X = K
    # if K is at-the-money-forward
    if abs(F - K) < 1e-12:
        numer1 = (((1 - beta)**2)/24)*alpha*alpha/(F**(2 - 2*beta))
        numer2 = 0.25*rho*beta*nu*alpha/(F**(1 - beta))
        numer3 = ((2 - 3*rho*rho)/24)*nu*nu
        VolAtm = alpha*(1 + (numer1 + numer2 + numer3)*T)/(F**(1-beta))
        sabrsigma = VolAtm
    else:
        z = (nu/alpha)*((F*X)**(0.5*(1-beta)))*np.log(F/X)
        zhi = np.log((((1 - 2*rho*z + z*z)**0.5) + z - rho)/(1 - rho))
        numer1 = (((1 - beta)**2)/24)*((alpha*alpha)/((F*X)**(1 - beta)))
        numer2 = 0.25*rho*beta*nu*alpha/((F*X)**((1 - beta)/2))
        numer3 = ((2 - 3*rho*rho)/24)*nu*nu
        numer = alpha*(1 + (numer1 + numer2 + numer3)*T)*z
        denom1 = ((1 - beta)**2/24)*(np.log(F/X))**2
        denom2 = (((1 - beta)**4)/1920)*((np.log(F/X))**4)
        denom = ((F*X)**((1 - beta)/2))*(1 + denom1 + denom2)*zhi
        sabrsigma = numer/denom

    return sabrsigma
```

>Step3: Estimate best parameters and calculate implied vol in SABR

```python
def sabr_vol(imp_df, S, r, T, strikes):
    F1 = S * np.exp(r*T)
    beta = 0.7   
    #sarb to fit market vol
    initialGuess = [0.02, 0.2, 0.1]
    res = least_squares(lambda x: sabrcalibration(x,
                                              imp_df['strike'],
                                              imp_df['impliedvol'],
                                              F1,
                                              T, beta),
                        initialGuess)
    alpha1 = res.x[0]
    rho1 = res.x[1]
    nu1 = res.x[2]

    print('Calibrated SABR model parameters: alpha = %.3f, beta = %.1f, rho = %.3f, nu = %.3f' % (alpha1, beta, rho1, nu1))




#calculate sarb using the fitting parameters
    sabrvols = []
    for K in strikes:
        sabrvols.append(SABR(F1, K, T, alpha1, beta, rho1, nu1))
    return sabrvols, alpha1, rho1, nu1
```

## The effect of parameters(rho, nu)--SABR

```python
rho = np.arange(-0.5,1,0.5)
rho = pd.DataFrame(rho,columns = ["rho"])

change_rho_of_sabrvols = []
for i in rho.index:
    rho_values = []
    for K in strikes3:    #spx maturity3 as example
        rho_values.append(SABR(Sx, K, T3, SABR_spx3[1], 0.7, rho["rho"].iloc[i], SABR_spx3[3]))
    change_rho_of_sabrvols.append(rho_values)
    
```

```python
nu = np.arange(0.1,0.7,0.2)
nu = pd.DataFrame(nu,columns = ["nu"])

change_nu_of_sabrvols = []
for i in nu.index:
    nu_values = []
    for K in strikes1:  #spx maturity1 as example
        nu_values.append(SABR(Sx, K, T1, SABR_spx1[1], 0.7, SABR_spx1[2], nu["nu"].iloc[i]))
    change_nu_of_sabrvols.append(nu_values)
    
```
