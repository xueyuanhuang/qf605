```python
ois = pd.read_excel("IR Data.xlsx", sheet_name = 'OIS', usecols = ["Tenor", "Product", "Rate"])
ois
```

```python
#tenors = ['6m', '1y', '2y', '3y', '4y', '5y', '7y', '10y', '15y', '20y', '30y']
tenors = ois["Tenor"]

# Define a function to convert tenor strings to days
def tenor_to_days(tenor):
    if 'm' in tenor:
        return int(tenor.replace('m', '')) * 30
    elif 'y' in tenor:
        return int(tenor.replace('y', '')) * 360
    else:
        return None

def tenor_to_years(tenor):
    if 'm' in tenor:
        return int(tenor.replace('m', '')) / 12
    elif 'y' in tenor:
        return int(tenor.replace('y', ''))
    else:
        return None

# Convert each tenor to days
ois["days"] = [tenor_to_days(tenor) for tenor in tenors]
ois["years"] = [tenor_to_years(tenor) for tenor in tenors]
ois
```

## D_6m

```python
from scipy.optimize import fsolve
import numpy as np

# Define the function that will represent the equation to solve
# We have to rearrange the equation to be in the form of f(x) = 0 for fsolve

#equation
#0.5*0.0025*D_6m = D_6m*((1 + f0/360)**180 - 1)
#D_6m = 1/(1 + f0/360)**180
def equations(vars):
    f0, D_6m = vars
    eq1 = D_6m - (1 / ((1 + f0/360) ** 180))
    eq2 = 0.5 * 0.0025 * D_6m - (D_6m * ((1 + f0/360) ** 180 - 1))
    return [eq1, eq2]

# Initial guess for f0 and D_6m
initial_guess = [0.01, 1]

# Use fsolve to find the solution
solution = fsolve(equations, initial_guess)

f0 = solution[0]
D_6m = solution[1]

f0, D_6m
```

## D_1y

```python
#equation
#0.003*D_1y = D_1y*(((1 + f0/360)**180)*(1 + f1/360)**180) - 1)
#D_1y = 1/((1 + f0/360)**180)*(1 + f1/360)**180))
def equations(vars):
    f1, D_1y = vars
    eq1 = D_1y - 1/(((1 + f0/360)**180)*((1 + f1/360)**180))
    eq2 = 0.003*D_1y - D_1y*(((1 + f0/360)**180) * ((1 + f1/360)**180) - 1)
    return [eq1, eq2]

# Initial guess for f0 and D_6m
initial_guess = [0.01, 1]

# Use fsolve to find the solution
solution = fsolve(equations, initial_guess)

f1 = solution[0]
D_1y = solution[1]

f1, D_1y
```

```python
# Given the known values for D_1y, f0, and f1, we can solve for D_2y and f2 using fsolve


# Define the equations based on the provided formulas
def equations(p):
    D_2y, f2 = p
    eq1 = 0.00325 * (D_1y + D_2y) - (D_1y * ((1 + f0/360) ** 180 * (1 + f1/360) ** 180 - 1) + D_2y * ((1 + f2/360) ** 360 - 1))
    eq2 = D_2y - (1 / ((1 + f0/360) ** 180 * (1 + f1/360) ** 180 * (1 + f2/360) ** 360))
    return (eq1, eq2)

# Initial guesses for D_2y and f2
initial_guesses = (0.995, 0.0035)

# Solve the equations
solution = fsolve(equations, initial_guesses)

D_2y, f2 = solution

D_2y, f2
```

```python
ois['discount_factor'] = 1/(1 + ois['Rate']/360)**ois['days']
ois
```

```python
def linear_interplolation(before_y, before_df, after_y, after_df, y):
    return before_df + ((y - before_y)/(after_y - before_y))*(after_df - before_df)

def curve(df, years):
    discount_factors = []
    for year in years:
        before = df[df['years'] <= year].iloc[-1]
        after = df[df['years'] >= year].iloc[0]
    
        if before['years'] == year:
            discount_factor = before["discount_factor"]
        elif after['years'] == year:
            discount_factor = after["discount_factor"]
        else:
            discount_factor = linear_interplolation(before['years'], before['discount_factor'], after['years'], after['discount_factor'], year)
    
        discount_factors.append(discount_factor)

    return pd.DataFrame({"years": years, "discount_factors": discount_factors})
```

```python
df = curve(ois[["years", "discount_factor"]], np.arange(0.5, 30.5, 0.5))
df
```

```python
plt.plot(df['years'], df['discount_factors'])
plt.show()
```
